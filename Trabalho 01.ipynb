__a) Descrição do experimento:__ foi utilizado o aplicativo iNVH, para Android, para medir as acelerações de vibração em um ventilador da marca "Britânia Turbo". O smarthphone foi encostado na frente do mesmo, sendo o ventilador então ligado na potência mínima. Isto foi necessário, para que fossem visíveis as formas de vibração. 

__b) Dados obtidos:__ para ler as informações foram utilizados os seguintes módulos do Python 3.7:

```python
import matplotlib.pyplot as plt
import numpy as np
from MRPy import MRPy
```

Os resultados obtidos são mostrados nas células seguintes:

```python
data = MRPy.from_file('ventilador', form='invh').zero_mean()
t    = data.t_axis()
plt.figure(1, figsize=(14, 8), clear=True)
  
for kX, acc in enumerate(data):
    
    plt.subplot(3, 1 ,kX+1)
    plt.plot(t, acc)

    plt.xlim(0, data.Td);   
    plt.ylim(-1.5, 1.5);
    plt.ylabel('acc{0} (m/s^2)'.format(kX+1)) 

    plt.grid(True) 

plt.xlabel('time (s)');
```
