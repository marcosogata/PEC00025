b) Foram utilizados também os seguintes módulos do Python 3.7:

```python
import matplotlib.pyplot as plt
import numpy as np
from MRPy import MRPy
```

Os resultados obtidos são mostrados nas células seguintes:

```python
data = MRPy.from_file('arcond', form='invh').zero_mean()
t    = data.t_axis()
plt.figure(1, figsize=(8, 8), clear=True)
  
for kX, acc in enumerate(data):
    
    plt.subplot(3, 1 ,kX+1)
    plt.plot(t, acc)

    plt.xlim(0, data.Td);   
    plt.ylim(-15, 15);
    plt.ylabel('acc{0} (m/s^2)'.format(kX+1)) 

    plt.grid(True) 

plt.xlabel('time (s)');
```
